````python
import pandas as pd
df = pd.read_csv(r"C:\Users\uSER\Downloads\daily_weather.csv")
df
````

````python
df.columns
````

````python
df.info()
````

````python
df.isnull().sum()
````

````python
df["air_pressure_9am"]=df["air_pressure_9am"].fillna(df["air_pressure_9am"].mean())
df["air_temp_9am"]=df["air_temp_9am"].fillna(df["air_temp_9am"].mean())
df["avg_wind_direction_9am"]=df["avg_wind_direction_9am"].fillna(df["avg_wind_direction_9am"].mean())
df["avg_wind_speed_9am"]=df["avg_wind_speed_9am"].fillna(df["avg_wind_speed_9am"].mean())
df["max_wind_direction_9am"]=df["max_wind_direction_9am"].fillna(df["max_wind_direction_9am"].mean())
df["max_wind_speed_9am"]=df["max_wind_speed_9am"].fillna(df["max_wind_speed_9am"].mean())
df["rain_accumulation_9am"]=df["rain_accumulation_9am"].fillna(df["rain_accumulation_9am"].mean())
df["rain_duration_9am"]=df["rain_duration_9am"].fillna(df["rain_duration_9am"].mean())
````

````python
df.isnull().sum()
````

````python
clean_df= df.copy()
clean_df["high_humidity_level"]=(df["relative_humidity_3pm"] > 24.99)*1
````

````python
clean_df
````

````python
y = clean_df[["high_humidity_level"]].copy()
y.head()
````

````python
clean_df[["relative_humidity_3pm"]].head()
````

````python
morning_features = ['air_pressure_9am', 'air_temp_9am', 'avg_wind_direction_9am',
       'avg_wind_speed_9am', 'max_wind_direction_9am', 'max_wind_speed_9am',
       'rain_accumulation_9am', 'rain_duration_9am']
````

````python
X = clean_df[morning_features].copy()
````

````python
X.columns
````

````python
X.isna().sum()
````

````python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
````

````python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.33, random_state = 0)
````

````python
humidity_classifier = DecisionTreeClassifier(max_leaf_nodes = 10, random_state = 0)
humidity_classifier.fit(X_train,y_train)
````

````python
type(humidity_classifier)
````

Prediction

````python
predictions = humidity_classifier.predict(X_test)
predictions[:10]
````

````python
y_test['high_humidity_level'][:10]
````

````python
accuracy_score(y_true = y_test, y_pred = predictions)
````
