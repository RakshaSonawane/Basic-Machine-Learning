```python
university_towns = []
with open(r"C:\Users\uSER\Downloads\university_towns.txt") as file:
    for line in file:
        if '[edit]' in line:
            state = line
        else:
            university_towns.append((state, line))
university_towns[:5]
```

```python
towns_df = pd.DataFrame(university_towns, columns=['State', 'RegionName'])
towns_df.head()
```

```python
def get_citystate(item):
    if ' (' in item:
        return item[:item.find(' (')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
```

```python
towns_df = towns_df.map(get_citystate)
```

```python
import pandas as pd
choco_df = pd.read_csv(r"C:\Users\uSER\Downloads\chocolate.csv")
choco_df
```

```python
col=['Unnamed: 0', 'ref', 'company', 'company_location', 'review_date',
       'country_of_bean_origin', 'specific_bean_origin_or_bar_name',
       'cocoa_percent', 'rating', 'counts_of_ingredients', 'beans',
       'cocoa_butter', 'vanilla', 'lecithin', 'salt', 'sugar',
       'sweetener_without_sugar']
choco_df = choco_df.drop(columns=col)
```

```python
choco_df.head()
```

```python
choco_df['first_taste'].value_counts()
```

```python
first_taste = choco_df['first_taste']
second_taste = choco_df['second_taste']
third_taste = choco_df['third_taste']
fourth_taste = choco_df['fourth_taste']
choco_df1 = pd.DataFrame({
    'first_taste': first_taste,
    'second_taste': second_taste,
    'third_taste': third_taste,
    'fourth_taste': fourth_taste,
})
```

```python
import numpy as np
new_df = pd.DataFrame()
melted_df = pd.melt(choco_df1, value_vars=['first_taste', 'second_taste', 'third_taste', 'fourth_taste'], var_name='original_column', value_name='combined')
choco_df['combined'] = melted_df[['combined']]
choco_df
```

```python
choco_df['combined'].value_counts().head(20)
```
