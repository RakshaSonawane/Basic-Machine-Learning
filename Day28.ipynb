```python
iqr_max = q3 + 1.5*iqr
iqr_min = q1 - 1.5*iqr
print(iqr_max, iqr_min)
filter_data = data[(data >= iqr_min) & (data <= iqr_max)]
filter_data
```

```python
df = pd.read_csv(r"C:\Users\uSER\Downloads\california_housing_test.csv")
df
```

```python
new_data = df.copy()
def outlier(col):
    q1 = np.percentile(col,25)
    q3 = np.percentile(col,75)
    iqr = q3-q1    
    iqr_max = q3 + 1.5*iqr
    iqr_min = q1 - 1.5*iqr
    filter_col = new_data[(col < iqr_min) | (col > iqr_max)].index
    for i in filter_col:
        new_data.drop(i, inplace=True)
    return new_data
```

```python
outlier(df['longitude'])
outlier(df['latitude'])
outlier(df['housing_median_age'])
outlier(df['total_rooms'])
outlier(df['total_bedrooms'])
outlier(df['population'])
outlier(df['households'])
outlier(df['median_income'])
outlier(df['median_house_value'])
```

```python
def l1_norm(col):
    return col/np.sum(np.abs(col))

def l2_norm(col):
    return col/np.sqrt(np.sum(np.square(col)))
```

```python
newdf_L1 = new_data.copy()
newdf_L1['longitude'] = l1_norm(new_data['longitude'])
newdf_L1['latitude'] = l1_norm(new_data['latitude'])
newdf_L1['housing_median_age'] = l1_norm(new_data['housing_median_age'])
newdf_L1['tatal_rooms'] = l1_norm(new_data['total_rooms'])
newdf_L1['total_bedrooms'] = l1_norm(new_data['total_bedrooms'])
newdf_L1['population'] = l1_norm(new_data['population'])
newdf_L1['households'] = l1_norm(new_data['households'])
newdf_L1['median_income'] = l1_norm(new_data['median_income'])
newdf_L1['median_house_value'] = l1_norm(new_data['median_house_value'])
newdf_L1
```

```python
newdf_L2 = new_data.copy()
newdf_L2['longitude'] = l2_norm(new_data['longitude'])
newdf_L2['latitude'] = l2_norm(new_data['latitude'])
newdf_L2['housing_median_age'] = l2_norm(new_data['housing_median_age'])
newdf_L2['tatal_rooms'] = l2_norm(new_data['total_rooms'])
newdf_L2['total_bedrooms'] = l2_norm(new_data['total_bedrooms'])
newdf_L2['population'] = l2_norm(new_data['population'])
newdf_L2['households'] = l2_norm(new_data['households'])
newdf_L2['median_income'] = l2_norm(new_data['median_income'])
newdf_L2['median_house_value'] = l2_norm(new_data['median_house_value'])
newdf_L2
```
