```python
n_dice=5
dice_faces={1,2,3,4,5,6}
event_space = set(pro(dice_faces, repeat= n_dice))
event_space
```

```python
len(event_space)
```

```python
favourable_outcome=[]
for outcome in event_space:
    d1,d2,d3,d4,d5 = outcome
    if (d1+d2+d3+d4+d5)%5==0 and (d1+d2+d3+d4+d5)%3 != 0:
        favourable_outcome.append(outcome)
favourable_outcome = set(favourable_outcome)
len(favourable_outcome)
```

```python
prob = len(favourable_outcome)/len(event_space)
print("The probability of getting a sum which is multiple of 5 but not of 3:", prob)
```

## Exponential Distribution

The time to failure of atomic system follows an exponential distribution with a mean time between Failure of 1000hour 

1. Calculate probability that systemm will fail before 1000 hour
2. Calculate probability that it will not fail up to 2000 hour
3. Calculate time by which 10% of system will fail(i.e. P 10 Life)

Since time to failure is 1000 hours, so lambda is 1/1000

```python
from scipy import stats
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
stats.expon.cdf(1000, loc=1/1000, scale=1000)
```

```python
1-stats.expon.cdf(2000, loc=1/1000, scale=1000)
```

```python
stats.expon.ppf(.1, loc=1/1000, scale=1000)
```

```python
pdf_df=pd.DataFrame({'success':range(0,5000,100), 'pdf': list(stats.expon.pdf(range(0,5000,100), loc=1/1000, scale=1000))})
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x=pdf_df.success, y= pdf_df.pdf, color='Green')
plt.xticks(rotation=90)
plt.xlabel('')
```
